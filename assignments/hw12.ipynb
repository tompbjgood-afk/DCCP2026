```python
import numpy as np
import pandas as pd


works = pd.DataFrame({
    '작가': ['김유정', '김유정', '김유정', '현진건', '현진건', '이상', '이상', '나도향', '나도향', '채만식', '이효석'],
    '작품': ['봄봄', '동백꽃', '만무방', '운수 좋은 날', 'B사감과 러브레터', '날개', '컨테', '벙어리 삼룡이', '물레방아', '레디메이드 인생', '메밀꽃 필 무렵'],
    '연도': [1935, 1936, 1934, 1924, 1925, 1936, 1937, 1925, 1925, 1934, np.nan],
    '글자수': [12500, 9800, 13600, 11200, 7400, 18300, 14100, 16800, 14500, 22000, 11800],
})


authors = pd.DataFrame({
    '작가': ['김유정', '현진건', '이상', '나도향', '채만식', '이효석', '염상섭'],
    '생년': [1908, 1900, 1910, 1902, 1902, 1907, 1897],
    '몰년': [1937, 1943, 1937, 1926, 1950, 1942, 1963],
})
```

```python
print(works.shape)
print(works.dtypes)
print(works.describe())
#pandas에서 int 열에 결측치가 포함되면 해당 열의 자료형이 float로 변환되기에 결측치가 있는 연도는 float형을 지닌다.
```

```python
print(works.isna().sum())
#각 열의 결측치를 카운팅하여 합한다. 연도 열에 1개의 결측치가 존재함을 확인할 수 있다.
```

```python
works2 = works.copy()
works2['연도'] = works2['연도'].fillna(1936).astype(int)
print(works2.dtypes)
print(works2.tail(3))
#기존 데이터를 안건들기 위해 복사본 만든 후 결측치를 1936으로 대체하고 열의 모든 값을 정수형으로 변환
```

```python
after_1930 = works2[works2['연도'] > 1930]
kim_or_lee = works2[works2['작가'].isin(['김유정', '이상'])]
print(after_1930)
print(kim_or_lee)
#1930년 이후에 발표된 작품 필터링하여 저장. isin을 활용해 김유정 또는 이상 작가의 작품만 저장
```

```python
def categorize(n: int) -> str:
    if n < 10000:
        return '짧음'
    elif n <= 15000: 
        return '보통'
    else:
        return '긴'

works2['분량'] = works2['글자수'].apply(categorize)
print(works2[['작품', '글자수', '분량']])
#분량을 정하는 함수를 정의하고 apply로 글자수에 따른 분량 추가
```

```python
sorted_works = works2.sort_values(by=['연도', '글자수'], ascending=[True, False])
print(sorted_works)
```

```python
author_history = works2.groupby('작가')['글자수'].agg(['mean', 'count', 'sum'])
print(author_history)
#groupby로 작가별로 글자수의 평균, 작품 수, 총 글자수를 계산하여 출력
```

```python
print(works2['분량'].value_counts())
```

```python
author_his2 = works2.groupby(['작가', '분량']).size()
print(author_his2)
#가장 많은 작품 수를 가진 조합은 (김유정, 보통)으로 2개이다. 이는 김유정 작가가 10,000자에서 15,000자 사이의 '보통' 크기 작품들을 선호하는 경향을 나타낸다.
```

```python
full = pd.merge(works2, authors, on='작가', how='left')
print(full)
print(full.shape)
#authors 표에는 존재하지만 works2 표에는 없는 작가(예: 염상섭)는 full 데이터프레임에 포함되지 않는다. 왜냐하면 how='left' 방식은 works2에 존재하는 키('작가')들을 기준으로 매칭을 수행하며, 오른쪽 표에만 존재하는 키값은 결합 결과에서 제외하기 때문이다.
```

```python
full['집필 당시 나이'] = full['연도'] - full['생년']
print(full[['작가', '작품', '연도', '생년', '집필 당시 나이']])
#집필 당시 나이를 계산하여 full 데이터 프레임에 추가한다.
```

```python
age = full.groupby('작가')['집필 당시 나이'].mean().sort_values()
print(age)
#가장 어린 나이에 작품을 발표한 작가는 나도향이며 평균 23살이다. 가장 늦은 나이에 작품을 발표한 작가는 채만식으로,평균 32살에 작품을 창작.
```
